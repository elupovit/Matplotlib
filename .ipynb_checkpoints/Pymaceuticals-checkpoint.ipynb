```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
clinical_to_load = "clinical_data.csv"
mouse_to_load = "mouse_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
clinical_df = pd.read_csv(clinical_to_load)
mouse_df = pd.read_csv(mouse_to_load)

# Combine the data into a single dataset
merged_data = pd.merge(clinical_df, mouse_df, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
merged_data
```

```python
#TUMOR GROWTH DURING TREATMENT 
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
vols_mean = merged_data.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
vols_mean_df = pd.DataFrame(vols_mean)

# Preview DataFrame
vols_mean_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
vol_std_err = merged_data.groupby(["Drug","Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
vol_std_err_df = pd.DataFrame(vol_std_err)

# Preview DataFrame
vol_std_err_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
d_size_over_time = vols_mean_df.reset_index().pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

d_size_over_time_std_err = vol_std_err_df.reset_index().pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

# Preview that Reformatting worked
d_size_over_time.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(d_size_over_time.index, d_size_over_time["Capomulin"], yerr=d_size_over_time_std_err["Capomulin"], color="b", marker=".", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(d_size_over_time.index, d_size_over_time["Infubinol"], yerr=d_size_over_time_std_err["Infubinol"], color="g", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(d_size_over_time.index, d_size_over_time["Ketapril"], yerr=d_size_over_time_std_err["Ketapril"], color="r", marker="v", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(d_size_over_time.index, d_size_over_time["Naftisol"], yerr=d_size_over_time_std_err["Naftisol"], color="c", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(d_size_over_time.index, d_size_over_time["Placebo"], yerr=d_size_over_time_std_err["Placebo"], color="m", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(d_size_over_time.index, d_size_over_time["Propriva"], yerr=d_size_over_time_std_err["Propriva"], color="y", marker="p", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(d_size_over_time.index, d_size_over_time["Ramicane"], yerr=d_size_over_time_std_err["Ramicane"], color="k", marker="*", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(d_size_over_time.index, d_size_over_time["Stelasyn"], yerr=d_size_over_time_std_err["Stelasyn"], color="orange", marker="h", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(d_size_over_time.index, d_size_over_time["Zoniferol"], yerr=d_size_over_time_std_err["Zoniferol"], color="purple", marker="x", markersize=5, linestyle="dashed", linewidth=0.50)


plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.grid(True)
plt.legend(("Capomulin","Infubinol","Ketapril","Naftisol","Placebo","Propriva","Ramicane","Stelasyn","Zoniferol"), loc="best", fontsize="small", fancybox=True)

#Save the figure
plt.savefig("Fig1_TumorGrowth.png")

#Show figure
plt.show()
```

```python
#METASTATIC TUMOR RESPONSE
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_mean =  merged_data.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]

# Convert to DataFrame
met_mean_df = pd.DataFrame(met_mean)

# Preview DataFrame
met_mean_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_std_err = merged_data.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]

# Convert to DataFrame
met_std_err_df = pd.DataFrame(met_std_err)

# Preview DataFrame
met_std_err_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_mean_df = met_mean_df.reset_index().pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

met_std_err_df = met_std_err_df.reset_index().pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

# Preview that Reformatting worked
met_mean_df.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(met_mean_df.index, met_mean_df["Capomulin"], yerr=met_std_err_df["Capomulin"], color="b", marker=".", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_mean_df.index, met_mean_df["Ceftamin"], yerr=met_std_err_df["Ceftamin"], color="g", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_mean_df.index, met_mean_df["Infubinol"], yerr=met_std_err_df["Infubinol"], color="r", marker="v", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_mean_df.index, met_mean_df["Ketapril"], yerr=met_std_err_df["Ketapril"], color="c", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_mean_df.index, met_mean_df["Naftisol"], yerr=met_std_err_df["Naftisol"], color="purple", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_mean_df.index, met_mean_df["Placebo"], yerr=met_std_err_df["Placebo"], color="y", marker="p", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_mean_df.index, met_mean_df["Propriva"], yerr=met_std_err_df["Propriva"], color="k", marker="h", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_mean_df.index, met_mean_df["Ramicane"], yerr=met_std_err_df["Ramicane"], color="orange", marker="+", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_mean_df.index, met_mean_df["Stelasyn"], yerr=met_std_err_df["Stelasyn"], color="pink", marker="x", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(met_mean_df.index, met_mean_df["Zoniferol"], yerr=met_std_err_df["Zoniferol"], color="brown", marker="*", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Metastatic Spread During Treatment")
plt.ylabel("Met. Sites")
plt.xlabel("Treatment Duration (Days)")
plt.grid(True)
plt.legend(("Capomulin","Infubinol","Ketapril","Naftisol","Placebo","Propriva","Ramicane","Stelasyn","Zoniferol"), loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("Fig2_MetSpread.png")

# Show the Figure
```

```python
#SURVIVAL RATE
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = merged_data.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].count()

# Convert to DataFrame
mice_count = pd.DataFrame(mice_count)

# Preview DataFrame
mice_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count = mice_count.reset_index().pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

# Preview the Data Frame
mice_count.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(((mice_count["Capomulin"] * 100)/ 25), color="b", marker=".", markersize=5, linewidth=0.5)
plt.plot(((mice_count["Ceftamin"] * 100)/ 25), color="g", marker="o", markersize=5, linewidth=0.5)
plt.plot(((mice_count["Infubinol"] * 100)/ 25), color="r", marker="v", markersize=5, linewidth=0.5)
plt.plot(((mice_count["Ketapril"] * 100)/ 25), color="c", marker="^", markersize=5, linewidth=0.5)
plt.plot(((mice_count["Naftisol"] * 100)/ 25), color="purple", marker="s", markersize=5, linewidth=0.5)
plt.plot(((mice_count["Placebo"] * 100)/ 25), color="y", marker="p", markersize=5, linewidth=0.5)
plt.plot(((mice_count["Propriva"] * 100)/ 25), color="k", marker="h", markersize=5, linewidth=0.5)
plt.plot(((mice_count["Ramicane"] * 100)/ 25), color="orange", marker="+", markersize=5, linewidth=0.5)
plt.plot(((mice_count["Stelasyn"] * 100)/ 25), color="pink", marker="x", markersize=5, linewidth=0.5)
plt.plot(((mice_count["Zoniferol"] * 100)/ 25), color="brown", marker="*", markersize=5, linewidth=0.5)

plt.title("Mouse Survival During Treatment")
plt.ylabel("Mice Survival (% Total)")
plt.xlabel("Time (Days)")
plt.grid(True)
plt.legend(("Capomulin","Infubinol","Ketapril","Naftisol","Placebo","Propriva","Ramicane","Stelasyn","Zoniferol"), loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("Fig3_Survival.png")

# Show the Figure
plt.show()
```

```python
#SUMMARY BAR GRAPH
# Calculate the percent changes for each drug
tum_per_d = (((d_size_over_time.iloc[-1]-d_size_over_time.iloc[0])/d_size_over_time.iloc[0])*100)
# Display the data to confirm
tum_per_d
```

```python
#Separate drug and percent changes into lists
drug = list(tum_per_d.index)
percent_change = list(tum_per_d.values)

#Associate colors to values in percent_change
colors = ['r' if d < 0 else 'g' for d in percent_change]

#Set x axis as array of tum_per_d values
x_axis = np.arange(len(tum_per_d))
```

```python
#Create bar chart
plt.bar(x_axis, tum_per_d, color = colors, align="center")

#Ticks
#Set xticks as values in x_axis
ticks = [x for x in x_axis]
#Plot ticks and rotate to properly visualize 
plt.xticks(ticks, tum_per_d.index, rotation=45)

#Set x and y limits
plt.xlim(0, len(x_axis))
plt.ylim(-30, (max(tum_per_d)+10))

#Title, x label, y label, establish grid
plt.title("Percentage Size Change Over Time")
plt.xlabel("Drugs")
plt.ylabel("Total % Volume Change")
plt.grid(True, lw=2, ls="--", color="grey")
```
